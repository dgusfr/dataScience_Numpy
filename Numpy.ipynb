## Regressão Linear

O coeficiente angular (𝑎) abaixo é usado para determinar a inclinação da reta de regressão linear, permitindo identificar a relação entre variáveis independentes e dependentes nos dados. Isso facilita prever tendências ou valores futuros com base em padrões históricos.

image.png

𝑥: Valores da variável independente.

𝑦: Valores da variável dependente.

𝑛: Número total de pontos de dados.

**Explicação**

Essa fórmula encontra a inclinação da reta de melhor ajuste para os pontos
(𝑥,𝑦).

O coeficiente angular (𝑎) é a taxa de variação da variável dependente
𝑦 em relação à variável independente 𝑥.

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
X.shape
```

```python
(X**2).shape
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
```

Depois de calcular
𝑎, a equação da reta é dada por:

**𝑦 = 𝑎 ⋅ 𝑥 + 𝑏**

Onde
𝑏 é o coeficiente linear, calculado por:

**b= n∑(y) − a∑(x)**







```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(datas, Moscow)
plt.plot(datas, y)
```

## Para que serve a regressão linear

No nosso caso, a regressão linear serve para modelar a relação entre variáveis nos dados, permitindo prever valores futuros ou tendências com base no comportamento passado. Ela ajuda a identificar padrões e entender como uma variável depende de outra, sendo essencial para análises quantitativas e suporte à tomada de decisões baseadas em dados.

```python
plt.plot(datas, Moscow)
plt.plot(datas, y)
plt.plot(41.5, a*41.5+b,'*r')
plt.plot(100, a*100+b,'*r')
```

Esse código plota os dados reais de Moscow em função do tempo (datas) e sobrepõe a reta de regressão linear ajustada, representando a relação entre as variáveis. Isso visualiza tanto os dados originais quanto a previsão linear e pontos de interesse.

## Reprodutibilidade

Números pseudoaleatórios são sequências de números gerados por algoritmos que tentam imitar o comportamento de sequências aleatórias. Eles não são verdadeiramente aleatórios, mas parecem ser aleatórios o suficiente para muitas aplicações práticas.

Os algoritmos geradores de números pseudoaleatórios geralmente iniciam com uma semente (seed), que é um número inicial que é fornecido como entrada para o algoritmo. A partir da semente, o algoritmo usa operações matemáticas para gerar uma sequência de números aparentemente aleatórios. No entanto, como o algoritmo é determinístico, a sequência de números gerada sempre será a mesma para a mesma semente.

```python
np.random.uniform(low=0.10,high=0.90,size=100)
```

```python
np.random.seed(16)
np.random.uniform(low=0.10,high=0.90,size=100)
```
